```python
prep = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(), cat_cols)
    ]
)

pipeline = Pipeline(steps=[
    ('preprocessor', prep),
    ('classifier', LogisticRegressionCV(cv=5, Cs=[100, 10, 1, 0.1, 0.01, 0.001], scoring='roc_auc'))
])
```

## Применение градиентного бустинга (2 балла)

```python
fpr, tpr, thresholds = roc_curve(y_valid, y_pred_prob)

roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='#6A7B73', lw=2, label=f'ROC curve (area = {roc_auc:.4f}')
plt.plot([0, 1], [0, 1], color='#E78F63', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

Попробуем подобрать гиперпараметры через optuna (а че так долго делается тооооо 10 минут 34 секундыыыыыыы)5469673643531326406.jpg

```python
def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 100, 1500, step=100),
        'depth': trial.suggest_int('depth', 3, 10),
        'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 1e-2, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, step=0.005),
        'grow_policy': trial.suggest_categorical('grow_policy', ['SymmetricTree', 'Lossguide']),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.5, 1.0),
        'random_seed': 42
    }

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    auc_scores = []

    for train_idx, valid_idx in cv.split(data[X], data[y]):
        X_train, X_valid = data[X].iloc[train_idx], data[X].iloc[valid_idx]
        y_train, y_valid = data[y].iloc[train_idx], data[y].iloc[valid_idx]

        model = CatBoostClassifier(
            **params,
            cat_features=cat_cols,
            eval_metric='AUC',
            verbose=0
        )
        model.fit(X_train, y_train, eval_set=(X_valid, y_valid), early_stopping_rounds=50)

        y_pred_prob = model.predict_proba(X_valid)[:, 1]
        auc = roc_auc_score(y_valid, y_pred_prob)
        auc_scores.append(auc)

    return np.mean(auc_scores)

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

print("Best parameters:", study.best_params)
print("Best ROC-AUC:", study.best_value)
```

Выпишите, какое лучшее качество и с какими параметрами вам удалось получить

Best parameters: {'iterations': 1200, 'depth': 3, 'l2_leaf_reg': 0.16318667194936673, 'learning_rate': 0.055, 'grow_policy': 'Lossguide', 'bagging_temperature': 0.22655207253147464, 'colsample_bylevel': 0.8235205771209145}
Best ROC-AUC: 0.8493416582401805

```python
best_params = study.best_params

final_model = CatBoostClassifier(
    **best_params,
    cat_features=cat_cols,
    eval_metric='AUC',
    random_seed=42

)

final_model.fit(X_train, y_train, eval_set=(X_valid, y_valid), early_stopping_rounds=50, verbose=0)

y_pred_prob_final_catboost = final_model.predict_proba(X_valid)[:, 1]

final_roc_auc = roc_auc_score(y_valid, y_pred_prob_final_catboost)
print(f'roc-auc: {final_roc_auc:.4f}')
```
